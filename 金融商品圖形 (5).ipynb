```python
#Just for investpy package
x=datetime.now()
from_date='01/06/2019'
to_date=x.strftime('%d/%m/%y20')
data=ip.get_index_historical_data(index='Dow 30',country='united states',from_date=from_date, to_date=to_date)#['Close']
def plot_k_dow_jones(data):
    #x=datetime.now()
    #from_date='01/12/2019'
    #to_date=x.strftime('%d/%m/%y20')
    datac=data['Close']
    datac=pd.DataFrame(datac)
    datac.rename(columns={'Close':'close'}, inplace=True)
    #return djic
    djima5=abstract.SMA(datac,10)
    djima20=abstract.SMA(datac,20)
    djima45=abstract.SMA(datac,45)
    djima60=abstract.SMA(datac,60)
    djima240=abstract.SMA(datac,240)
    data.columns=['open','high','low','close','x','y']
    data=data.drop(columns =['x', 'y']) 
    datakd=abstract.STOCH(data,fastk_period=9, slowk_period=3,slowd_period=3)
    #畫k線圖資料預處理
    data=data.reset_index()
    data.columns=['Datetime','Open','High','Low','Close']
    data['Datetime']=mdates.date2num(data['Datetime']) #要用matplotlib 畫k線圖 必須把日期轉換成float
    #畫k線圖
    length=len(data['Datetime'].values[60:])
    figure=plt.figure(facecolor='white',figsize=(20,10))
    ax1=plt.subplot2grid((5,4),(0,0),rowspan=4,colspan=4,facecolor='white')
    candlestick_ohlc(ax1, data.values,width=0.6,colorup='red',colordown='green')
    ax1.xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d'))
    ax1.plot(data['Datetime'].values[-length:],djima45[-length:],'blue',label='ma45', linewidth=1.5)
    ax1.legend(loc='best',ncol=5)
    #ax1.grid(True,color='black',axis='x')
    plt.ylabel('stock price',fontsize=18)
    plt.suptitle('Dow_jones_index',color='black',fontsize=28)
    # 
    ##畫kd值
    ax2=plt.subplot2grid((5,4),(4,0),sharex=ax1,rowspan=5,colspan=4,facecolor='white')
    ax2.plot(data['Datetime'].values[-length:],datakd['slowk'].values[-length:],color='red')
    ax2.plot(data['Datetime'].values[-length:],datakd['slowd'].values[-length:],color='blue')
    ax2.grid(True,color='black',axis='y')
    plt.ylabel('KD Value',fontsize=18)
    plt.yticks([20,80])
    plt.savefig(fname = "道瓊指數指數.png", format = "png")
```
